```python
trainData = graphlab.SFrame('image_train_data/')
testData = graphlab.SFrame('image_test_data/')
```

```python
trainData['label'].sketch_summary()
```

```python
itemSet = set()
for i in range(len(trainData)):
    word = trainData[i]['label']
    itemSet.add(word)

trainDataDict = {}

for item in itemSet:
    #print item
    data = trainData[trainData['label'] == item]
    trainDataDict[item] = data

testDataDict = {}

for item in itemSet:
    #print item
    data = testData[testData['label'] == item]
    testDataDict[item] = data
    
#birdData = trainData[trainData['label'] == 'bird']
#catData = trainData[trainData['label'] == 'cat']
#dogData = trainData[trainData['label'] == 'dog']
#automobileData = trainData[trainData['label'] == 'automobile']
```

```python
#dataDict['dog']
```

```python
knnModelDict = {}
for item in itemSet:
    knnModel = graphlab.nearest_neighbors.create(trainDataDict[item], features = ['deep_features'], label = 'id')
    knnModelDict[item] = knnModel
```

```python
dogDistSf = graphlab.SFrame({})

for item in itemSet:
    sf = knnModelDict[item].query(testDataDict['dog'], k = 1)['distance']
    dogDistSf.add_column(sf, name = item)
```

```python
dogDistSf
```

```python
length = len(dogDistSf['dog'])
lt = [True] * length
for i in range(length):
    if (dogDistSf[i]['dog'] > dogDistSf[i]['cat'] or 
        dogDistSf[i]['dog'] > dogDistSf[i]['bird'] or 
        dogDistSf[i]['dog'] > dogDistSf[i]['automobile']):
        lt[i] = False
dogDistSf['True False'] = graphlab.SArray(lt)

dogDistSf
```

```python
cCnt = dogDistSf['True False'].sum()
fCnt = length - cCnt
print "True = %d, Flase = %d\n" %(cCnt, fCnt)
```
