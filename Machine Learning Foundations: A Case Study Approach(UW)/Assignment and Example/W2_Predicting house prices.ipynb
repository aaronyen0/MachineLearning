```python
import graphlab
```

# Load some house sales data

```python
sales = graphlab.SFrame('.\\Week 2\\home_data.gl/')
```

```python
sales
```

```python
sales.show()
```

# Exploring the data for housing sales

```python
graphlab.canvas.set_target('ipynb')
sales.show(view = "Scatter Plot", x = "sqft_living", y = "price")
```

# Create a simple regression model of sqfr_living to price

```python
training_data, test_data = sales.random_split(0.8, seed = 123)
```

```python
training_data.show()
```

## Build the regression model

```python
sqft_model = graphlab.linear_regression.create(training_data, target = 'price', features = ['sqft_living'])
```

```python
print test_data['price'].mean()
```

```python
sqft_model.evaluate(test_data)
```

# Let's show what our predictions look like

```python
import matplotlib.pyplot as Plt
%matplotlib inline
```

```python
Plt.plot(test_data['sqft_living'], test_data['price'], '.',
        test_data['sqft_living'], sqft_model.predict(test_data),'-')
```

```python
sqft_model.get("coefficients")
```
